# Operadores relacionais

- \> maior
- < menor
- == igual
- \>= maior igual
- <= menor igual
- != diferente

```python
10 > 20
```

```python
10 < 20
```

```python
nome_aluno = 'vinicius'
nome_professor = 'Vinicius'
```

```python
nome_aluno == nome_professor
```

```python
10 != 10
```

# Operadores lógicos

- AND (e)
- OR (ou)
- NOT (negação)

## AND

operador_AND.JPG

```python
(10 > 20) and (20 > 30)
```

```python
(10 < 20) and (20 < 30)
```

## OR

operador_OR.JPG

```python
(1 < 2) or (3 > 6)
```

```python
'Vinicius' == 'Vinicius' or 'Python' != 'python'
```

## NOT

operador_NOT.JPG

```python
not True
```

```python
not False
```
